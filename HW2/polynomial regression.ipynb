```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt
```

```python
training_set = pd.read_csv('train_set.csv')
test_set = pd.read_csv('test_set.csv')
```

```python
training_set.head(5)
```

```python
training_set.plot(kind="scatter", x="x", y="target")
```

```python
X=np.array(training_set['x']).reshape(-1, 1)
y=training_set['target']
poly_features = PolynomialFeatures(degree=5, include_bias=False)
X = poly_features.fit_transform(X)
```

```python
X_test = np.array(test_set['x']).reshape(-1, 1)
y_test=test_set['target']
X_test_poly=poly_features.fit_transform(X_test)
```

```python
def poly_reg(n_samples = 200):
  X_train = X[:n_samples]
  y_train = y[:n_samples]
  lin_reg = LinearRegression()
  lin_reg.fit(X_train, y_train)
  y_predictions = lin_reg.predict(X_test_poly)
  lin_mse = mean_squared_error(y_test, y_predictions)
  return y_predictions, lin_mse,lin_reg
```

```python
y_prediction_10, lin_mse_10,reg_10 = poly_reg(n_samples = 10)
y_prediction_25, lin_mse_25,reg_25 = poly_reg(n_samples = 25)
y_prediction_50, lin_mse_50,reg_50 = poly_reg(n_samples = 50)
y_prediction_100, lin_mse_100,reg_100 = poly_reg(n_samples = 100)
y_prediction_200, lin_mse_200,reg_200 = poly_reg(n_samples = 200)
```

```python
fig, axs = plt.subplots(2, 3)
fig.set_figwidth(20)
fig.set_figheight(15)
axs[0, 0].plot(y_test, 'r', label='Y',alpha=0.6)
axs[0, 0].plot(y_prediction_10, 'b', label='Y_pred',alpha=0.6)
axs[0, 0].set_title("number of samples : 10")
axs[0, 1].plot(y_test, 'r', label='Y',alpha=0.6)
axs[0, 1].plot(y_prediction_25, 'b', label='Y_pred',alpha=0.6)
axs[0, 1].set_title("number of samples : 25")
axs[0, 2].plot(y_test, 'r', label='Y',alpha=0.6)
axs[0, 2].plot(y_prediction_50, 'b', label='Y_pred',alpha=0.6)
axs[0, 2].set_title("number of samples : 50")
axs[1, 0].plot(y_test, 'r', label='Y',alpha=0.6)
axs[1, 0].plot(y_prediction_100, 'b', label='Y_pred',alpha=0.6)
axs[1, 0].set_title("number of samples : 100")
axs[1, 1].plot(y_test, 'r', label='Y',alpha=0.6)
axs[1, 1].plot(y_prediction_200, 'b', label='Y_pred',alpha=0.6)
axs[1, 1].set_title("number of samples : 200")
fig.legend([axs[0,0]], labels=['Y','Y_pred'],loc='upper center', shadow=True, fontsize='x-large')
```

```python
lin_mse = [lin_mse_10,lin_mse_25,lin_mse_50,lin_mse_100,lin_mse_200]
```
