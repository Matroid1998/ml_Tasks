```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('2_train_set.csv')
```

```python
df
```

```python
train_lb = df.drop(columns = 'target',axis = 1)
y_train = df['target']
```

```python
for i in range(10):
  plt.imshow(np.array(train_lb.iloc[i]).reshape(8,8),cmap='gray')
  print(df.iloc[i,-1])
  plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
```

```python
train_lb_std = std.fit_transform(train_lb)
```

# Logistic Regression


```python
df_test = pd.read_csv('2_test_set.csv')
```

```python
test=df_test.drop(columns = 'target',axis = 1)
y_test = df_test['target']
```

```python
test_std = std.transform(test)
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import GridSearchCV, RepeatedStratifiedKFold
cv = RepeatedStratifiedKFold(n_splits=5, n_repeats=20, random_state=42)

sgd = SGDClassifier(loss='log')
params_grid = [{'alpha': [0.00001,0.0001,0.0005,0.001,0.005,0.01,0.1,1]}]
sgd_gs = GridSearchCV(sgd,params_grid, cv = cv,scoring = 'accuracy')
```

```python
sgd_gs.fit(train_lb_std,y_train)
```

```python
sgd_gs.best_params_
```

```python
sgd_gs.best_score_
```

```python
sgd_best = SGDClassifier(loss = 'log',alpha = 0.1)
sgd_best.fit(train_lb_std,y_train)
```

```python
y_predicted = sgd_best.predict(test_std)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_predicted)
```

```python
from sklearn.cluster import KMeans
k = 40
```

## PCA

```python
train_unlb = pd.read_csv('2_unlabeled.csv')
```

```python
train_unlb
```

```python
plt.imshow(np.array(train_unlb.iloc[0]).reshape(8,8),cmap = 'gray')
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(train_unlb)
```

```python
pca_df = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
```

# Kmeans

```python
std2 = StandardScaler()
train_unlb_std = std2.fit_transform(train_unlb)
```

```python
kmeans = KMeans(n_clusters=k,init = 'k-means++',random_state=42)
```

```python
clusters_label= kmeans.labels_
plt.scatter(x=pca_df['principal component 1'], y=pca_df["principal component 2"],c = clusters_label)
plt.show()
```

```python
X_cluster_dist = kmeans.fit_transform(train_unlb_std)
```

```python
X_cluster_dist.shape
```

```python
X_cluster_dist[0]
```

```python
center_digit_idx = np.argmin(X_cluster_dist, axis=0)
```

```python
center_digits = train_unlb_std[center_digit_idx]
```

```python
center_digits.shape
```

```python
plt.figure(figsize=(20,20))
for i in range(40):
  plt.subplot(10, 4, i+1)
  plt.imshow(std2.inverse_transform(center_digits[i].reshape(1,-1)).reshape(8,8),cmap='gray')
```

```python
manual_labels = [7,1,9,0,
                 0,3,1,6,
                 2,4,4,6,
                 3,8,2,4,
                 1,4,5,1,
                 4,4,7,4,
                 8,9,6,7,
                 8,4,4,0,
                 5,1,2,1,
                 2,1,7,5
                 ]
```

```python
updated_labels = np.empty(len(train_unlb_std), dtype=np.int32)
```

```python
for i in range(k):
  updated_labels[kmeans.labels_==i] = manual_labels[i]
```

```python
updated_labels
```

```python
semi_supervised_data = pd.concat([train_unlb,pd.DataFrame(updated_labels,index = train_unlb.index,columns =['target'])],axis = 1)
```

```python
semi_supervised_data
```

```python
j=1
for i in range(10):
  s = np.array(semi_supervised_data[semi_supervised_data['target']==i][:10]).reshape(10,-1)
  for img in s :
    plt.subplot(10,10,j)
    plt.imshow(img[:-1].reshape(8,8),cmap = 'gray')
    j+=1
```

```python
train = pd.concat([df,semi_supervised_data],axis = 0)
```

```python
train
```

```python
X_train = train.drop(columns = ['target'],axis=1)
y = train['target']
```
