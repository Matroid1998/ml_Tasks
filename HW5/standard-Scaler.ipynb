```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
train = pd.read_csv('train_set.csv')
test = pd.read_csv('test_set.csv')
```

```python
train.head(5)
```

```python
train.describe()
```

```python
train.info()
```

```python
X_train = train.drop(columns=['income'])
y_train = train['income']
```

```python
num_columns = X_train.select_dtypes(include=[np.number]).columns
std = StandardScaler()
```

```python
X_train[num_columns] = pd.DataFrame(std.fit_transform(X_train[num_columns]),columns=num_columns)
```

```python
test[num_columns] = pd.DataFrame(std.transform(test[num_columns]),columns=num_columns)
```

```python
X_train
```

```python
train = pd.concat([X_train,y_train],axis=1)
```

```python
train.to_csv('train_std.csv')
```

```python
test.to_csv('test_std.csv')
```
